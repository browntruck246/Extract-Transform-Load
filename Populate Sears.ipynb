🧱 SQL: Create a Customer Table

```python
%sql
CREATE DATABASE IF NOT EXISTS `Sears`;

CREATE TABLE `Sears`.`Customers` (
    Customer_ID INT PRIMARY KEY,
    First_Name VARCHAR(50),
    Last_Name VARCHAR(50),
    Email VARCHAR(100),
    Phone VARCHAR(20),
    Address VARCHAR(150),
    City VARCHAR(50),
    State VARCHAR(50),
    Zip_Code VARCHAR(10)
);
```
